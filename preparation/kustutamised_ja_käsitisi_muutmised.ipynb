```python
from estnltk.vabamorf.morf import synthesize
def correct_word_form(word, vabamorf):
  word = word.split(".")
  word = word[0]
  sõna = ""
  for w in word.split():
    synthesized = synthesize(w, vabamorf)
    if synthesized:
        sõna += synthesized[0] + " "
    else:
        return

  if sõna == "":
    return word
  return sõna.strip()
```

```python
# leiame 100 kõige populaarsemat lemmat:
most_popular_lemmas = dict()
for col in df_loomad.columns:
  question = Text(col).tag_layer().morph_analysis.lemma
  for lemma in question:
    if lemma[0] not in most_popular_lemmas:
      most_popular_lemmas[lemma[0]] = 1
    else :
      most_popular_lemmas[lemma[0]] += 1
# Sorteerime sageduse järgi (kahanevalt)
sorted_lemmas = sorted(most_popular_lemmas.items(), key=lambda x: x[1], reverse=True)

# Väljastame top 100 lemmat
for lemma, freq in sorted_lemmas[:200]:
    print(f"{lemma}: {freq}")
```

```python
# lisame populaarsete lemmade põhjal veel üleüldisemaid küsimusi

categories = {
    "color": [
        "valge", "must", "pruun", "pruunikas", "punane", "punakas", "punakaspruun", "hall", "hallikaspruun", "tume", "tumepruun"
    ],
    "location": [
        "Aafrika", "Lõuna-Ameerika", "Põhja-Ameerika", "Eesti", "Aasia", "Ameerika",
        "Kagu-Aasia", "Euroopa", "Euraasia", "Austraalia", "Kesk-Ameerika", "Lõuna-Aasia"
    ],
    "body_part": ["sarv", "nokk", "saba", "sulestik"],
    "hair": ["karv", "karvastik", "karvkate"]
}
category_questions = {
    "color": lambda word: f"Kas see loom on {word}?",
    "location": lambda word: f"Kas see loom paikneb {correct_word_form(word, 'sg in')}?",
    "body_part": lambda word: f"Kas sellel loomal on {word}?",
    "hair": lambda word: f"Kas sellel loomal on karvastik?"
}

def get_matching_words_for_lemma(df, target_lemma):
    matching_words = set()
    for col in df.columns:
        if target_lemma.lower() in col.lower():
            words = df.loc[df[col] == 1, "Sõna"]
            matching_words.update(words)
    return matching_words

def add_general_questions_by_lemma(df, categories, category_questions):
    for category, lemmas in categories.items():
        for lemma in lemmas:
            question = category_questions[category](lemma)
            matching_words = get_matching_words_for_lemma(df, lemma)
            if question not in df.columns:
                df[question] = 0
            df.loc[df["Sõna"].isin(matching_words), question] = 1
    return df

df_loomad = add_general_questions_by_lemma(df_loomad, categories, category_questions)
```

```python
# 1. Filtreerime tähtsusega read
taimed_filtered = df_taimed[df_taimed["tähtsus"] == 1]
loomad_filtered = df_loomad[df_loomad["tähtsus"] == 1]

# 2. Määrame, milliseid veerge mitte eemaldada (ID, metaandmed jms)
columns_to_exclude = ['Sõna', 'tähtsus', 'Synonyms']  # lisa siia, mida veel alles tahad jätta

# 3. Leiame veerud, millel pole ainult 0-id (või stringina "0")
def get_nonzero_columns(df):
    columns_to_check = [col for col in df.columns if col not in columns_to_exclude]
    numeric_df = df[columns_to_check].apply(pd.to_numeric, errors='coerce').fillna(0)
    nonzero_columns = numeric_df.loc[:, numeric_df.ne(0).any()].columns
    return nonzero_columns.tolist()

taimed_nonzero_cols = get_nonzero_columns(taimed_filtered)
loomad_nonzero_cols = get_nonzero_columns(loomad_filtered)

# 4. Kombineerime need veerud ja lisame kohustuslikud veerud juurde
final_columns = list(set(taimed_nonzero_cols + loomad_nonzero_cols + columns_to_exclude))

# 5. Lõikame veerud välja ja ühendame andmestikud
combined_df = pd.concat([
    taimed_filtered[taimed_filtered.columns.intersection(final_columns)],
    loomad_filtered[loomad_filtered.columns.intersection(final_columns)]
], ignore_index=True)
```
