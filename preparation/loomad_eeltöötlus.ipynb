````python
from estnltk.vabamorf.morf import synthesize
def correct_word_form(word, vabamorf):
  word = word.split(".")
  word = word[0]
  sõna = ""
  for w in word.split():
    synthesized = synthesize(w, vabamorf)
    if synthesized:
        sõna += synthesized[0] + " "
    else:
        return

  if sõna == "":
    return word
  return sõna.strip()
````

### 2.2 definitsiooni põhjal küsimused

````python
# Täpsemad küsimused -- kasutame ainult definitsioone

# (examples pole mõtet kasutada, sest nendest ei saa sisukaid küsimusi,
# nt (['Varblased tunglesid saiapalukese ümber.'] example)": {'varblane'},
# (['Nägime kardinali, kelle pea oli helepunane.'] example)": {'kardinal'}, (mõni hea aga pole väärt seda tööd))
````

````python
import re
from collections import defaultdict

def generate_specific_questions(sentence):
    # Eemaldab ebavajalikud sõnad, leiab POS-märgendid ja teeb küsimused

    # eemaldame sulud ja sulgude sees olevad sõnad
    sentence = re.sub(r'\([^)]*\)', '', sentence)

    new_sentence = ""
    words_with_pos = []

    text = Text(sentence).tag_layer()

    for word in text.morph_analysis:
        wordPOS = word.partofspeech[0]
        words_with_pos.append((word.text, wordPOS))  # Salvestame nii sõna kui ka POS-i

    pos = [pos for _, pos in words_with_pos]  # Loome POS-ide listi
    seq_str = " ".join(pos)  # Muudame listi stringiks

    # 1. variant: "keegi, kes VERB..." -> "kas see loom VERB"
    if re.search(r"\bP Z P V\b", seq_str):
        remaining_words = [word for word, wordPOS in words_with_pos if wordPOS not in ["P", "Z", "P"]]
        result = f"Kas see loom {' '.join(remaining_words)}?"
        return result

    # 2. variant: asesõna + VERB -> asesõna asemele läheb 'loom' ('Kas see loom on ta on hundisuurune; tüvepikkus on 76–102 cm, saba pikkus 31–41 cm, õlakõrgus 60 cm ja mass 16–23 kg? {'hüään'})
    if re.search(r"^P V\b", seq_str):
        remaining_words = [word for word, wordPOS in words_with_pos[1:]]
        result = f"Kas see loom {' '.join(remaining_words)}?"
        return result
    else:
      result = f"Kas see loom on {sentence.strip()}?"
      return result
````

````python
def pair_questions_answers(category):
    # küsimused ja vastused kokku
    definition_word = defaultdict(set)

    for words in category.closure("hyponym"):

        if words.definition:
          specific_questions = generate_specific_questions(words.definition)
          #specific_questions = specific_questions + " (" + str(words.definition)+ ")" # et näha tervet words.defi
          definition_word[specific_questions].add(correct_word_form(words.name , "sg n"))

    return definition_word
````

### 2.3 Wordneti teised seosed

1. Kui ma hetkel võtan mingi mõiste ja saan teada, millega ta seotud on, siis kas ma saan kindel olla, et selle mõiste alammõisted on ka nende sõnadega seotud

2. Olen saanud kätte sõnad ja nende teised seosed. Nüüd tuleks teha eraldi kontroll, et ma ei lisaks küsimusi, mis on juba olemas ala siidisabale 'kas on laululind?', sest eelduslikult võiks selline küsimus juba olemas olla. Samas tuleks üle kontrollida kas ikka on

3. millised küsimusi täpselt koostada

4. olenevalt seose tüübist tuleb ka küsimus moodustada (vb tuleb mõni seos veel eemaldada)



````python
def otherRelations(synsets):
    word_info = []
    for synset in synsets:
        relations = {}

        for relation in wn.all_relation_types():
            if relation not in ["hyponym", "hypernym", "domain_topic", "antonym", "similar"]:
                related_synsets = synset.get_related_synset(relation)
                if related_synsets:
                    relations[relation] = [str(r_synset.lemmas[0]) for r_synset in related_synsets]

        if relations:  # Ainult siis, kui on seoseid
            word_info.append([synset, relations])

    return word_info
````

````python
# vaatame milliseid seoseid oleks mõistlik kasutada

all_related_words_by_type = {}

for mõiste, seosed in otherRelations:
    for relation_type, relations in seosed.items():
        if relation_type not in all_related_words_by_type:
            all_related_words_by_type[relation_type] = set()  # Kasutame hulka, et vältida kordusi
        all_related_words_by_type[relation_type].update(relations)  # Lisame seotud sõnad hulka

for relation_type in all_related_words_by_type:
    all_related_words_by_type[relation_type] = list(all_related_words_by_type[relation_type])

for relation_type, words in all_related_words_by_type.items():
    print(f"{relation_type}: {words}")
````

````python
# kasutame edaspidi ainult mõistlikuid seosed (holo_member, mero_member, involved_location, agent, holo_location)
valid_relation_types = {"holo_member", "mero_member", "involved_location", "agent", "holo_location"}

filtered_otherRelations = []

for mõiste, seosed in otherRelations:
    filtered_seosed = {relation_type: relations for relation_type, relations in seosed.items() if relation_type in valid_relation_types}
    if filtered_seosed:
        filtered_otherRelations.append([mõiste, filtered_seosed])

# filtreeritud ja sobivad seosed
otherRelations = filtered_otherRelations
````

````python
# Teha seoste sõnadest küsimused (kasutame ainult neid seoseid, sest need on kõige mõistlikumad:
# holo_member, mero_member, involved_location, agent (tegusõnad), holo_location)

# word_and_relations = {'holo_part': ['kanakasvatus'], 'mero_part': ['kanamuna']}
# Põhjendus: # akki state_of eemaldada: sõna 'koduveis' seos 'vasikas' ({'state_of': ['vasikas']
# Sõna be_in_state eemaldada 'isatiiger' seos 'isane' ({'be_in_state': ['isane']}) esines küsimuses 'Kas see loom on isane vasikas?', aga pole lisatud {'härgvasikas'}

def generateOtherRelationsQuestion(relation, relation_type):
  question = ""

  if relation_type in  ["holo_member", "mero_member"]:
    question = f"Kas tegu on {correct_word_form(relation, 'sg kom')}?"
  elif relation_type in  ['involved_location', 'holo_location']:
    question = f"Kas seda looma võib leiduda {correct_word_form(relation, 'sg in')}?"
  elif relation_type in  ["agent"]:
    relation_pos = Text(relation).tag_layer().morph_analysis.partofspeech[0]
    if relation_pos[0] == "V": # võtame ainult tegusõnad
      question = f"Kas see loom {correct_word_form(relation, 'b')}?"
  return question
````

````python
# tahan lisada:
# küsimused, mis on juba olemas, aga kus on üks mõiste puudu
# küsimused, mida ei ole veel olemas

def otherRelationsQuestion(otherRelations, yleyldised_küsimused_vastused):
    questions = set()
    seosed_küsimused_vastused = defaultdict(set)

    for küsimus, mõisted in yleyldised_küsimused_vastused.items():
        lemmad = Text(küsimus).tag_layer().words.lemma
        lemmad_list = [lemma[0] for lemma in lemmad]

        for mõiste, seosed in otherRelations:
            mõiste_nimi = correct_word_form(mõiste.name, "sg n")

            for key, words in seosed.items():
                for word in words:
                    if word not in lemmad_list:
                        # Uus küsimus, mida veel ei ole olemas
                        question = generateOtherRelationsQuestion(word, key)
                        questions.add(question)
                        seosed_küsimused_vastused.setdefault(question, set())
                        seosed_küsimused_vastused[question].add(mõiste_nimi)

                        # Lisa alammõisted, kui need on olemas
                        for alam_mõiste in mõiste.closure("hyponym"):
                            seosed_küsimused_vastused[question].add(correct_word_form(alam_mõiste.name, "sg n"))

                    # Küsimus on juba olemas, kontrollime, kas tuleb mõiste juurde lisada
                    elif word in lemmad_list and mõiste_nimi not in mõisted:
                      yleyldised_küsimused_vastused[küsimus].add(mõiste_nimi)

                    else:
                      continue

    return seosed_küsimused_vastused, yleyldised_küsimused_vastused

````

````python
def lemmatiseerimine(tekst, stopwords = sagedussõnastik):
    lemmatiseeritud_tekst = []
    # Lemmatiseerimine
    sentences = Text(tekst).tag_layer().sentences
    for sentence in sentences:
      lemmad = sentence.words.lemma

      # Eemaldame stopp-sõnad
      lemmad_wo_stopwords = [lemma[0] for lemma in lemmad if lemma[0] not in stopwords]

      # Tagastame puhastatud tekstina, liites sõnad tühikuga
      lemmatiseeritud_lause = ' '.join(lemmad_wo_stopwords)
      lemmatiseeritud_tekst.append(lemmatiseeritud_lause)

    return lemmatiseeritud_tekst
````
